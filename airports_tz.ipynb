```python
import pandas as pd
```

# Airports in Tanzania
List of airports in Tanzania, with latitude and longitude. 
* The Humanitarian Data Exchange: https://data.humdata.org/dataset/ourairports-tza
* Unverified community data from http://ourairports.com/countries/TZ/

```python
humdata_tz_airports = pd.read_csv('HumData_tz-airports.csv')
opendata_airports = pd.read_csv('OpenData_airports.csv')
```

```python
# humdata_tz_airports.head()
# opendata_airports.head()
```

```python
humdata_tz_airports = humdata_tz_airports.copy(deep=True)
humdata_tz_airports = humdata_tz_airports.drop(
    columns=['continent', 'scheduled_service', 'home_link', 'wikipedia_link', 'score', 'last_updated']).copy(deep=True)
humdata_tz_airports
```

```python
opendata_tz_airports = opendata_airports[opendata_airports['iso_country'] == 'TZ'].copy(deep=True).reset_index()
opendata_tz_airports = opendata_tz_airports.drop(
    columns=['index', 'continent', 'iso_country', 'scheduled_service', 'local_code', 'home_link', 'wikipedia_link']).copy(deep=True)
opendata_tz_airports
```

```python
humdata_tz_airports.to_csv('humdata_tz_airports_clean.csv')
opendata_tz_airports.to_csv('opendata_tz_airports_clean.csv')
```
